```python
# Initialize the dictionary to store matrices
matrices = {"matrix 1": None, "matrix 2": None, "result matrix": None}

# Function to collect a matrix and store it in the dictionary
def collect_matrix(matrix_name):
    print(f"Enter {matrix_name}:")
    rows = int(input("Enter the number of rows: "))
    cols = int(input("Enter the number of columns: "))
    
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = float(input(f"Enter element at row {i+1}, col {j+1}: "))
            row.append(element)
        matrix.append(row)
    
    matrices[matrix_name] = matrix
    print(f"{matrix_name} has been entered successfully.")

# Function to perform matrix multiplication and store the result
def multiply_matrices():
    matrix1 = matrices["matrix 1"]
    matrix2 = matrices["matrix 2"]
    
    if matrix1 is None or matrix2 is None:
        print("Both matrices must be entered first.")
        return
    
    if len(matrix1[0]) != len(matrix2):
        print("Matrix multiplication is not possible. Number of columns in matrix 1 must equal the number of rows in matrix 2.")
        return
    
    result_matrix = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix2[0])):
            cell_value = sum(matrix1[i][k] * matrix2[k][j] for k in range(len(matrix1[0])))
            row.append(cell_value)
        result_matrix.append(row)
    
    matrices["result matrix"] = result_matrix
    print("Matrix multiplication completed successfully.")
def export_result_to_file(filename):
    result_matrix = matrices["result matrix"]
    
    if result_matrix is None:
        print("Result matrix is not available.")
        return
    
    with open(filename, "w") as file:
        file.write("Result Matrix:\n")
        for row in result_matrix:
            file.write(" ".join(map(str, row)) + "\n")
    
    print(f"Result matrix has been exported to {filename}.")

# Main menu loop
while True:
    print("\nOptions:")
    print("1. Enter Matrix 1")
    print("2. Enter Matrix 2")
    print("3. Calculate")
    print("4. Export Result to File")
    print("5. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == "1":
        collect_matrix("matrix 1")
    elif choice == "2":
        collect_matrix("matrix 2")
    elif choice == "3":
        multiply_matrices()
    elif choice == "4":
        filename = input("Enter the filename to export the result: ")
        export_result_to_file(filename)
    elif choice == "5":
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```
