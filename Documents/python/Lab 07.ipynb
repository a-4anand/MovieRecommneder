```python
import random

# Define the menu items and their prices
menu = [
    ["Rice", 30],
    ["Tea", 10],
    ["Pizza", 50],
    ["Burger", 40],
    ["Salad", 35],
    # Add more items as needed
]

# Initialize the list to store orders
orders = []

# Function to create a new order
def create_order():
    order_id = generate_order_id()
    items = []
    print("\nMenu:")
    for i, item in enumerate(menu):
        print(f"{i+1}. {item[0]} - ${item[1]}")
    while True:
        try:
            choice = int(input("Enter item number (0 to finish): "))
            if choice == 0:
                break
            elif choice < 1 or choice > len(menu):
                print("Invalid item number. Please try again.")
                continue
            quantity = int(input("Enter quantity: "))
            items.append({"itemnumber": choice, "quantity": quantity})
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    orders.append({"Orderid": order_id, "items": items})
    print(f"Order {order_id} created successfully!")

# Function to generate a unique order ID
def generate_order_id():
    order_id = ''.join(str(random.randint(0, 9)) for _ in range(5))
    return int(order_id)

# Function to calculate the total bill for an order
def calculate_order_total(order):
    total = 0
    for item in order["items"]:
        item_number = item["itemnumber"]
        quantity = item["quantity"]
        total += menu[item_number - 1][1] * quantity
    return total

# Function to view orders and calculate bills
def view_orders():
    if not orders:
        print("No orders found.")
        return
    for order in orders:
        order_id = order["Orderid"]
        total = calculate_order_total(order)
        print(f"Order {order_id}:")
        for item in order["items"]:
            item_number = item["itemnumber"]
            quantity = item["quantity"]
            item_name = menu[item_number - 1][0]
            item_price = menu[item_number - 1][1]
            print(f"{item_name} x{quantity} - ${item_price * quantity}")
        print(f"Total: ${total}\n")

# Main menu loop
while True:
    print("Options:")
    print("1. Create new order")
    print("2. View Orders")
    print("3. Exit")
    choice = input("Enter your choice: ")
    if choice == '1':
        create_order()
    elif choice == '2':
        view_orders()
    elif choice == '3':
        print("Exiting the application. Thank you!")
        break
    else:
        print("Invalid choice. Please try again.")
```
